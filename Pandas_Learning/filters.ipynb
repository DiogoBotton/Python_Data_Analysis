Nesse momento, o time de ML deseja avaliar apenas imóveis em cenários específicos, como:

1. Apartamentos que possuem 1 quarto e aluguel menor que R$ 1200;

2. Apartamentos que possuem pelo menos 2 quartos, aluguel menor que R$ 3000 e área maior que 70 m².

Com base nisso, precisamos aplicar os filtros necessários para obter essas informações específicas.

```python
import pandas as pd

df = pd.read_csv('aluguel_apartamentos.csv', sep=',')
```

Poderiamos realizar estes filtros simplesmente com o método "query", porém usaremos outros meios para exemplificar várias formas de fazer buscas.

```python
df.query('Quartos == 1 & Valor < 1200')
```

```python
selecao1 = df['Quartos'] == 1
selecao2 = df['Valor'] < 1200

# Atentar-se que é necessário colocar entre parenteses os dois filtros e o operador condicional "and"(&)
df[(selecao1) & (selecao2)]
```

```python
# A mesma lógica, porém em apenas uma única linha:
df[(df['Quartos'] == 1) & (df['Valor'] < 1200)]
```

O próximo filtro, número 2:

```python
# Os operadores condicionais devem ser & ou | obrigatóriamente (como and ou or não funcionam)
df[(df['Quartos'] >= 2) & (df['Valor'] < 3000) & (df['Area'] > 70)] 
```
