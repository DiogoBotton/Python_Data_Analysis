```python
import numpy as np
import matplotlib.pyplot as plt

# Desenhando um circulo no gráfico

# Gerando valores de x de -1 a 1
x = np.arange(-1, 1.01, 0.0001)

y1 = np.sqrt(1-x**2)
y2 = -np.sqrt(1-x**2)

# Plotar o gráfico com a parte de cima (positiva) e de baixo (negativa) do círculo
plt.plot(x, y1, 'r')
plt.plot(x, y2, 'r')

# Adiciona alguns títulos
plt.title("Círculo")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")

plt.show()
```
